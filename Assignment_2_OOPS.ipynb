```python
# 10. How can you have multiple constructors in a Python class? Explain with an example.
# there are two main approaches :
    # 1.Method Overloading:
        # . python dos not support method overloading in the same way that some other languages 
        # do (eg. java or c++), where you can define multiple methods with same name but different prameter lists
        # . however , you can achieve a similar effect by defining multiple methods with different names to ac as construtors ,each of which performs different type of object initialization
        
# 2.Using '@classmethod' and factory methods:
    # . you can use class methods (methods decorated with '@classmethod') to create alternative
    # constructers often refered to as factory methods
    # These factory methods create and return instances of the class with specific initializations based on the arguments provided.
    #Factory methods ar commonly used to provide multiple ways of creating objects with different attributes
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    @classmethod
    def create_square(cls,side_length):
        #factory method to create a square (a special case of a rectangle)
        return cls(side_length,side_length)
    @classmethod
    def create_default_rectangle(cls):
        ##factory method to create a default rectagle with a length of 1 and width of 1
        return cls(1,1)
    
#using primary constructor
rec1 = Rectangle(5,3)
print(f"Rectangle 1 area:{rec1.area()}")

##using factory method to create a square
square = Rectangle.create_square(4)
print(f"Square area :{square.area()}")

##using a factory method to create a defaut rectangle 
default_rect = Rectangle.create_default_rectangle()
print("default Rectangle area :",default_rect.area())
```

```python
# 1. What is inheritance in Python? Explain its significance in object-oriented programming.
# in python is a core concept of object-oriented programming (oop) that allows a new class to be based on an exising class
# in inheritance , the new class (called the 'subclass' or 'derived class')inherits attributes and behviours from an existing class(called the base class) 
# in python you create an inheritance relationship by specifying the base class in the definition of the subclass
# Concept of Inheritance:
# 1). Inheritance allow you to define a new class by reusing and extending an existing class
# 2).The subclass inherits the attributes (data) and methods (functions) of the base class, and it can also have its own attributes and methods.

# Significance in OOP:
# Code Reusability
# Hierarchy and Classification
# Polymorphism
# Extensibility
# Abstraction
class vehicle:
    def __init__(self,make,model):
        self.make = make
        self.model = model
    def start_engine(self):
        print("Engine started")
class car(vehicle):
    def __init__(self,make,model,color):
        super().__init__(make,model)
        self.color = color
    def drive(self):
        print(f"{self.color} car is driving")
my_car = car("toyota","camry","blue")
my_car.start_engine()
my_car.drive()
```

```python
# 4. Explain the concept of method overriding in inheritance. Provide a practical example.
# Method overriding:
    # it is a funcdamental concept in inheritance in object oriented programming 
    # it occurs when a subclass provides  a specific implementation of a method that is already defined in its parent class
    # this allows the subclass to replace or extend the behaviourof the inherited method
    # method overriding is a way to achieve polymorphism,where objects of different classes can repond to a common method in away that is appropriate for their specific class
#how method overriding works:
    # the parent class defines a method
    # the child class redefines the same method with the same name and parameters , providing a specific implementation
# When you call the overridden method on an instance of the child class, the subclass's implementation is executed, not the parent class's.
class shape:
    def area(self):
        pass
class Circle(shape):
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.14* self.radius * self.radius
class Square(shape):
    def __init__(self,side):
        self.side = side
    def area(self):
        return self.side*self.side
circle = Circle(23)
square = Square(34)
print(f"Circle Area:{circle.area()}")  # Calls the overridden method in Circle
print(f"Square Area: {square.area()}") # Calls the overridden method in Square
```

```python
# 5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.
# ans-> you can access the method and attributes of a parent class from a child class using the "super()" function .
# the super() function allow you to call methods and access attributes of the parent class within the child class
# this often used to extend or override the behaviour of the parent classes methods while still utilizing their functionality
class parent:
    def __init__(self,name):
        self.name = name
    def say_hello(self):
        print(f"hello, I am {self.name}")
class child(parent):
    def __init__(self,name,age):
        super().__init__(name)  # Call the parent class constructor to initialize 'name'
        self.age = age
        
    def say_hello(self):
        super().say_hello()  # Call the 'say_hello' method of the parent class
        print(f"i am {self.age} years old")
child = child("rahul",34)
child.say_hello()
```

```python
# 6. Discuss the use of the `super()` function in Python inheritance. When and why is it used? Provide an example.
# ans:- 
# The 'super()' function in python is used in the context of inheritance to access and call methods or attibutes of a parent from a child 
# it often used when you want to extend or overrid the behaviour of a method define in th eparent class while still using the parent class implementation 'super()' is particularly helpful in case of method overriding to ensure that you dont lose the functionally provided by the parent class

# how and why 'super()' is used :
# 1)Accessing Parent Class Methods: You can use super() to call a method defined in the parent class from the child class. 
    # This allows you to execute the parent class's behavior and then extend or modify it as needed.

# 2)Preventing Code Duplication: By using super(), you avoid code duplication. Instead of rewriting the entire method from the parent class in the child class, you can call the parent class's method and add only the specific functionality relevant to the child class.

#3)Method Overriding: When you override a method in the child class, you can use super() to call the parent class's version of the method, thereby combining the behaviors of both classes.
#4)Maintaining a Clean and Readable Codebase: super() enhances code readability by making it clear that you're accessing the parent class's method. This is particularly useful in complex inheritance hierarchies.
class parent:
    def greet(self):
        return "Hello from the parent class"
class child(parent):
    def greet(self):
        parent_greeting = super().greet()
        return f"{parent_greeting} and hi from the child class"
child1 = child()
child1.greet()
```

```python
# What is the purpose of the `issubclass()` function in Python? Provide an example.
# The 'issubclass()' function in python is used to determine if a class is a subclass of another class
# it allows you to check the inheritance relationship between two classe
# the 'issubclass()' function is particularly useful when you want to verify if one class is derived from another class in an object-oriented context

# issubclass(class, classinfo) takes two arguments: class is the class you want to check, and classinfo is either a class or a tuple of classes to compare against.
# If class is a subclass of any of the classes specified in classinfo, issubclass() returns True. Otherwise, it returns False.
class Animal:
    pass
class Mammal(Animal):
    pass
class Dog(Animal):
    pass
class Bird(Animal):
    pass

# using issubclass() to check inheritance
print(issubclass(Mammal,Animal))  # True, Mammal is a subclass of Animal
print(issubclass(Dog,Animal)) # True, Dog is a subclass of Animal
print(issubclass(Dog,Mammal)) # True, Dog is a subclass of Mammal
print(issubclass(Animal,Bird)) # False, Animal is not a subclass of Bird
    
```

```python
# 12. Explain the concept of property decorators in Python and how they relate to encapsulation.
# property decorater in python are amechnism that allows you to define methods that act like attributes, providing a way to control access, modification and validation of class attributes
# they are used to implement the concept of encopsulation by providing controlled access to an enable you to hde the internal implementation details of an attribut while exposing a clean and controlled interface for accessing and modifying it

##How property decorators work and thir relation to encapsulation

# 1. Getter Method: you can use  the @%property decorater to define a method that is accessed like an attribute
# >this method is used to get the value of an attribute , and it is executed when you access the attribute
# >it allos you to perform any necessary logic before returning the attribute's value .
# >This is particularly useful for encapsulating the attribute's internal

class myclass:
    def __init__(self):
        self._value = 65 #Private attribute
        
    @property
    def value(self):
        ##Getter method for encapsulation
        return self._value
obj = myclass()
print(obj.value) ##accessing value using the property

# 2. Setter Method: You can use the @attribute_name_setter decorater to define a method a method that is excutd when you assign a value to the attribute
# this allows you to control the asignment and validation of the attributes value, enhansing encapsulation

class myclass:
    def __init__(self):
        self.value = 34
    @property
    def value(self):
        return self.value
    @value.setter
    def value(self,new_val):
        if new_val >=0:
            self._value = new_val
obj = myclass()
obj.value = 42 ##setting value using the property
# print(obj.value)

# Deleter method: You can use the @attribute_name.deleter decorator to define a method that is executed when you delete the attribute. 
# This can be used to implement custom logic when deleting an attribute, further enhancing encapsulation.

class myclass:
    def __init__(self):
        self.value = 34
    @property
    def value(self):
        return self.value
    @value.setter
    def value(self,new_val):
        if new_val >=0:
            self._value = new_val
    @value.deleter
    def value(self):
        #Deleter the value attribute
        print("Deleting The value attribute")
        del self._value
        
obj = myclass()
del obj.value ##deleting the value
```

```python
# 15. Describe the concept of operator overloading in Python and how it relates to polymorphism. Provide
# examples using operators like `+` and `*`.

# >>Operator Overloading in python allows you to define or redefine the behaviour of operators for user-defined classes 
# >This means you can make objects of your class work with the standard 
# >python operators like '+','-',"-"etc
# >operator overloading is clodely related to polymorphism because it enables objects of different types to work seamlessly with operators,
# provide a unified interface.

# here some example of method overloading the '+' Operator
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Point):
            # Adding two Point objects
            return Point(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            # Adding a Point object and a scalar
            return Point(self.x + other, self.y + other)
        else:
            raise ValueError("Unsupported operand type")

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

# Example usage:

point1 = Point(1, 2)
point2 = Point(3, 4)
result_point = point1 + point2
print(result_point)  

result_point_with_scalar = point1 + 5
print(result_point_with_scalar) 


# Example 2: Overloading the * Operator
class Vector:
    def __init__(self, values):
        self.values = values

    def __mul__(self, scalar):
        # Multiplying a Vector by a scalar
        if isinstance(scalar, (int, float)):
            return Vector([value * scalar for value in self.values])
        else:
            raise ValueError("Unsupported operand type")

    def __repr__(self):
        return f"Vector({self.values})"

# Example usage:

vector1 = Vector([1, 2, 3])
result_vector = vector1 * 2
print(result_vector)  # Output: Vector([2, 4, 6])

result_vector_with_float = vector1 * 1.5
print(result_vector_with_float)  # Output: Vector([1.5, 3.0, 4.5])
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @property
    @abstractmethod
    def area(self):
        pass

    @property
    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self._width = width
        self._height = height

    @property
    def area(self):
        return self._width * self._height

    @property
    def perimeter(self):
        return 2 * (self._width + self._height)

class Circle(Shape):
    def __init__(self, radius):
        self._radius = radius

    @property
    def area(self):
        import math
        return math.pi * self._radius ** 2

    @property
    def perimeter(self):
        import math
        return 2 * math.pi * self._radius

# Example usage
rectangle = Rectangle(5, 10)
circle = Circle(7)

print(f"Rectangle Area: {rectangle.area}")
print(f"Rectangle Perimeter: {rectangle.perimeter}")

print(f"Circle Area: {circle.area}")
print(f"Circle Perimeter: {circle.perimeter}")
```

```python
# 15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.

# >>Abstract Data Types (ADTs) are a concept in computer science that refers to a high-level description of a set of data and the operations that can performed a=on that data.
# >>ADTs provide a way to abstractly represent and interact with data structures, emphasizing what operation can be performed on the data rather than how those operations are implemented 
# >> This level of abstraction allows developers to use data structures witgout concerning themselves with the underlying details of their implementation

# here are keypoints about the concept of ADTs abd their role in achieving abstraction i python:

# 1) Abstraction:
    # ADTs encapsulate the essential properties of a data structure and hide the details of its implementation. 
    #This allows users to focus on what the data structure does rather than how it achieves its functionality.
    # Users interact with the ADT through a well-defined set of operations, 
    # without needing to know the internal workings of the data structure.

# 2)Interface vs. Implementation:
    # ADTs separate the interface (the set of operations and their specifications) from the implementation (how those operations are carried out)
    # This distinction is crucial for achieving abstraction.
    # For example, a stack ADT provides operations like push, pop, and is_empty. Users of the stack don't need to know whether the stack is implemented using arrays or linked lists.

# 3) Encapsulation:
    # ADTs often use encapsulation to bundle data and operations together. 
    # The internal details are hidden from the user, and access to the data is controlled through well-defined interfaces.
    # In Python, classes are commonly used to implement ADTs. The methods of the class define the operations of the ADT, and the attributes represent the data.
    
# Examples of ADTs:
    # Common examples of ADTs include stacks, queues, linked lists, trees, and dictionaries. Each ADT provides a set of operations that can be performed on the data it manages, abstracting away the details of how the data is stored and manipulated.
```

```python
from abc import ABC, abstractmethod

class LibraryItem(ABC):
    def __init__(self, title, author, item_id):
        self.title = title
        self.author = author
        self.item_id = item_id
        self.checked_out = False

    @abstractmethod
    def check_availability(self):
        pass

    @abstractmethod
    def display_info(self):
        pass

    def borrow_item(self):
        if not self.checked_out:
            self.checked_out = True
            print(f"{self.title} by {self.author} has been borrowed.")
        else:
            print(f"{self.title} is currently unavailable.")

    def return_item(self):
        if self.checked_out:
            self.checked_out = False
            print(f"{self.title} has been returned.")
        else:
            print(f"{self.title} was not checked out.")


class Book(LibraryItem):
    def __init__(self, title, author, item_id, num_pages):
        super().__init__(title, author, item_id)
        self.num_pages = num_pages

    def check_availability(self):
        return "Available" if not self.checked_out else "Checked Out"

    def display_info(self):
        print(f"Book ID: {self.item_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Number of Pages: {self.num_pages}")
        print(f"Availability: {self.check_availability()}")


# Example usage:

# Create a book instance
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "B001", 180)

# Display book information
book1.display_info()

# Borrow the book
book1.borrow_item()

# Try to borrow the book again (it's already checked out)
book1.borrow_item()

# Return the book
book1.return_item()

# Try to return the book again (it's already returned)
book1.return_item()
```

```python
# 8. Create a Python class for a computer system, using composition to represent components like CPU, RAM,
# and storage devices.
class CPU:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def process_data(self):
        print(f"{self.brand} CPU processing data at {self.speed} GHz")

class RAM:
    def __init__(self, capacity):
        self.capacity = capacity

    def store_data(self):
        print(f"RAM storing data with a capacity of {self.capacity} GB")

class Storage:
    def __init__(self, capacity, storage_type):
        self.capacity = capacity
        self.storage_type = storage_type

    def read_data(self):
        print(f"{self.storage_type} storage reading data with a capacity of {self.capacity} GB")

class ComputerSystem:
    def __init__(self, cpu, ram, storage):
        self.cpu = cpu
        self.ram = ram
        self.storage = storage

    def perform_computing(self):
        print("Performing computing tasks:")
        self.cpu.process_data()
        self.ram.store_data()
        self.storage.read_data()

# Example usage
if __name__ == "__main__":
    # Create components
    my_cpu = CPU(brand="Intel", speed=3.5)
    my_ram = RAM(capacity=16)
    my_storage = Storage(capacity=512, storage_type="SSD")

    # Create computer system using composition
    my_computer = ComputerSystem(cpu=my_cpu, ram=my_ram, storage=my_storage)

    # Perform computing tasks
    my_computer.perform_computing()
```

```python
# 9. Describe the concept of "delegation" in composition and how it simplifies the design of complex systems.
# >>Delegation in composition is a design principle in object-oriented programming where one class passes the responsibility of a certain task to another class.
# >> Instead of performing the task directly, the delegating class delegates the responsibility to another class, typically one it contains as a component
#>> This allows for a more modular and flexible design, simplifying the complexity of systems.


# Here's an explanation of how delegation works and its benefits:
# 1.Delegation Process:
    # >>In delegation, a class A delegates a specific task to a class B that it contains as a component.
    # >>Class A maintains a reference to an instance of class B and forwards the task to that instance.
    # >>Class B, as the delegate, is responsible for implementing the task.

# Example:
    # >>Consider a scenario where a TaskManager class needs to perform file operations. Instead of implementing file operations directly in the TaskManager class, it delegates this responsibility to a separate FileManager class.
class FileManager:
    def perform_file_operation(self):
        print("File operation performed by FileManager")

class TaskManager:
    def __init__(self, file_manager):
        self.file_manager = file_manager

    def execute_task(self):
        print("Executing task in TaskManager")
        self.file_manager.perform_file_operation()

# Example usage
if __name__ == "__main__":
    my_file_manager = FileManager()
    my_task_manager = TaskManager(file_manager=my_file_manager)

    my_task_manager.execute_task()
# In this example:
# The TaskManager class delegates the responsibility of file operations to the FileManager class.
# The TaskManager class contains an instance of the FileManager class and calls its perform_file_operation method.
```

```python
# 16. Create a Python class for a computer game character, using composition to represent attributes like
# weapons, armor, and inventory.
class Weapon():
    def __init__(self,name,damage):
        self.name = name
        self.damage = damage
        
class Armor:
    def __init__(self,name,defense):
        self.name = name
        self.defense = defense
        
class Inventary_item:
    def __init__(self,name,quantity):
        self.name = name
        self.quantity = quantity
class GameCharater:
    def __init__(self,name):
        self.name = name
        self.equiped_weapon = None
        self.equipment_armor = None
        self.inventary = []
    
    def equip_weapon(self,weapon):
        self.equiped_weapon = weapon
    def equip_armor(self,armor):
        self.equipment_armor = armor
    def add_to_inventary(self,item):
        self.inventary.append(item)
    def display_charcters(self):
        print("Character:",self.name)
        if self.equiped_weapon is not None:
            print("Equipped Weapon:", self.equiped_weapon.name, "Damage:", self.equiped_weapon.damage)
        if self.equipment_armor is not None:
            print("Equipped Armor:", self.equipment_armor.name, "Defense:", self.equipment_armor.defense)
        if not self.inventary:
            print("Inventory:")
            for item in self.inventary:
                print("-", item.name, "Quantity:", item.quantity)
            
```

```python
sword = Weapon(name="Sword",damage=10)
shield = Armor("shield",5)
health_posistion = Inventary_item(name="Health Position",quantity=3)

player_charater = GameCharater("Hero")
player_charater.equip_weapon(sword)
player_charater.equip_armor(shield)
player_charater.add_to_inventary(health_posistion)
player_charater.display_charcters()
```

```python
# 17. Describe the concept of "aggregation" in composition and how it differs from simple composition.
# >>Aggregation is a concept in object-oriented programming (OOP) and software design that involves creating relationships between classes to represent a "whole-part" relationship
# >>It is a form of composition where one class (the whole or container) contains objects of another class (the part or component).
# >>The key characteristic of aggregation is that the parts can exist independently of the whole.
# >>In aggregation, the relationship between the container class and the component class is often depicted as a "has-a" relationship.

# >>For example, a university class may have students, but those students can exist outside of the specific class they are enrolled in. If the class is disbanded, the students still exist.

# >>Here are some key points about aggregation:

# >>1.Independence of components: In aggregation, the components (parts) can exist independently of the container (whole).
# If the whole is destroyed, the parts can still exist.

# >>2.Multiplicity: Aggregation can involve a one-to-many relationship, meaning one container class can have multiple instances of the component class.

# >>3.Flexibility: Aggregation provides flexibility because the components can be shared among different containers or used in isolation.


# >>>Now, let's contrast aggregation with simple composition:

# >Composition: In simple composition, the components are strongly tied to the container, and their existence is dependent on the container.
# > If the container is destroyed, the components are also destroyed.
# >It represents a "owns-a" relationship.
# >For example, if a car is composed of an engine, the engine may cease to exist if the car is scrapped.
# >Aggregation vs. Composition: The main difference lies in the independence of components.
# >In aggregation, components can exist on their own, while in composition, components are part of the whole and cannot exist independently.
```

```python
# 18. Create a Python class for a house, using composition to represent rooms, furniture, and appliances.
class Furniture:
    def __init__(this,name):
        this.name = name
    def describe(this):
        return f"This is a {this.name}"
class Appliances:
    def __init__(this,name):
        this.name = name
    def describe(this):
        return f"This is a {this.name}"
class Room:
    def __init__(this,name):
        this.name = name
        this.furniture = []
        this.appliances = []
    def add_furniture(this,furniture):
        this.appliances.append(furniture)
    
    def add_application(this,appliances):
        this.appliances.append(appliances)
    
    def describe(this):
        furniture_description = "\n".join(item.describe() for item in this.furniture)
        appliances_description = "\n".join(item.describe() for item in this.appliances)
        return f"\nRoom : {this.name} \n Furniture:\n{furniture_description}\nAppliances:{appliances_description}"
class House:
    def __init__(this,name):
        this.name = name
        this.rooms = []
    def add_room(this,room):
        this.rooms.append(room)
    
    def describe(this):
        room_description = "\n".join(room.describe() for room in this.rooms)
        return f"House: {this.name}\nRooms:\n{room_description}"
living_room = Room("Living Room")
living_room.add_furniture(Furniture("Sofa"))
living_room.add_furniture(Furniture("Cofee Table"))
living_room.add_application(Appliances("Television"))

kitchen = Room("Kitchen")
kitchen.add_furniture(Furniture("Sofa"))
kitchen.add_furniture(Appliances("Refrigerator"))
kitchen.add_furniture(Appliances("Oven"))

my_house = House("Cozy Cottage")
my_house.add_room(living_room)
my_house.add_room(kitchen)

print(my_house.describe())
```

```python
# 20. Create a Python class for a social media application, using composition to represent users, posts, and
# comments.
class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

    def display(self):
        return f"{self.user.username}: {self.text}"

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display(self):
        post_info = f"{self.user.username} posted: {self.content}"
        comments_info = "\n".join(comment.display() for comment in self.comments)
        return f"{post_info}\nComments:\n{comments_info}"

class User:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def create_post(self, content):
        post = Post(self, content)
        self.posts.append(post)
        return post

    def comment_on_post(self, post, text):
        comment = Comment(self, text)
        post.add_comment(comment)
        return comment

    def display_posts(self):
        return "\n".join(post.display() for post in self.posts)

# Example usage:
user1 = User("Alice")
user2 = User("Bob")

post1 = user1.create_post("Hello, world!")
comment1 = user2.comment_on_post(post1, "Nice post, Alice!")
comment2 = user1.comment_on_post(post1, "Thanks, Bob!")

post2 = user2.create_post("Python is awesome!")
comment3 = user1.comment_on_post(post2, "I totally agree!")

# Displaying posts and comments
print(user1.display_posts())
print(user2.display_posts())
```
